Create a line plot showing the number of marriages and divorces per capita in the 
U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python

import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
details = pd.read_csv("us-marriages-divorces-1867-2014.csv")
margs_p_capita = details['Marriages_per_1000']
divrs_p_capita = details['Divorces_per_1000']
columns = margs_p_capita, divrs_p_capita
years = details['Year']
ind = ["margs_p_capita", "divrs_p_capita"]


x_details = range(1867, 1867 + details.shape[0])
fig, axis = plt.subplots()
for i in range(len(columns)):
    axis.plot(x_details, columns[i], label=ind[i])    
axis.set_xlabel('Year')
axis.set_ylabel('The number of Marriages and Divorces')
axis.legend()        
```
